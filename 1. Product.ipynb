### Pricing Analysis

```python
data.head()
```

```python
#6.What is the distribution of product prices?

plt.figure(figsize=(10,6))
sns.histplot(data = data,x = 'price',bins=30,kde=True,color = 'blue')

plt.title('Distribution of Product Price')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

```python
#7. What is the most expensive product, and what category does it belong to?

data.nlargest(1,'price')
```

```python
#8.What is the least expensive product, and what category does it belong to?

data.nsmallest(1,'price')
```

```python
#9. Are there significant price differences between categories?

data.groupby('category')['price'].describe()
```

```python
#10. Which product categories are most common?

category_counts = data['category'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='orange')
plt.title('Product Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

### Size and Calorie Analysis

```python
#18. What is the average calorie count for each category?

data.groupby('category')['calories'].mean().reset_index()
```

### Availability and Ratings

```python
#23. What percentage of products are currently available?

deva = len(data)

raju = data['availability'].sum()

percentage = (raju/deva) * 100

print(f"The percentage of products currently available is: {percentage:.2f}%")
```

```python
data.head()
```

```python
#24. What is the relationship between price and calories?

plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='calories', data=data, hue='category', palette='viridis')
plt.title('Price vs Calories by Category')
plt.xlabel('Price')
plt.ylabel('Calories')
plt.legend(title='Category')
plt.show()
```

```python
#25. What is the distribution of ratings across all products?

plt.figure(figsize=(10, 6))
sns.histplot(data['rating'], bins=20, kde=True, color='purple')

plt.title('Distribution of Ratings Across All Products', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```

```python
#26. Which category has the highest average rating?

deva = data.groupby('category')['rating'].mean()

word = deva.idxmax()
word1 = deva.max()

print(f"The category with the highest average rating is: {word}")
print(f"The highest average rating is: {word1:.2f}")
```

```python
#27. Are certain product sizes associated with higher ratings?

data.groupby('size')['rating'].mean().reset_index()
```

### Advanced and Cross-Category Analysis

```python
#28. Are there patterns in product popularity (based on stock quantity and ratings) by category or size? 

data.groupby(['category', 'size'])[['quantity', 'rating']].mean()
```

```python
#29. Is there any relationship between product flavors and their availability or rating?

data.groupby('flavors')[['availability', 'rating']].mean()
```

```python
#30. Which flavors are the most popular?

flavor_counts = data['flavors'].value_counts().head(10)

plt.figure(figsize=(10, 6))
flavor_counts.plot(kind='bar', color='green')
plt.title('Top 10 Most Popular Flavors')
plt.xlabel('Flavors')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#31. How is availability distributed across categories?

plt.figure(figsize=(12, 6))
sns.countplot(x='category', hue='availability', data=data, palette='Set2')
plt.title('Availability by Product Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Availability', labels=['Unavailable', 'Available'])
plt.show()
```
