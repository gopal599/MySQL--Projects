```python
# Question 1: Total number of orders

total_orders = data['order_id'].nunique()
print("Total number of orders:", total_orders)
```

```python
# Question 2: Total budget across all orders

total_budget = data['budget'].sum()
print("Total budget:", total_budget)
```

```python
# Question 3: Number of unique customers

unique_customers = data['customer_id'].nunique()
print("Unique customers:", unique_customers)
```

```python
# Question 4: Orders of 'Delivery' vs 'Takeaway'

order_types = data['type_of_order'].value_counts()
print("Order types:\n", order_types)
```

```python
# Question 5: Average budget for all orders

avg_budget = data['budget'].mean()
print("Average budget:", avg_budget)
```

```python
# Question 6: Distribution of budgets by order type

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='budget', hue='type_of_order', multiple="stack", kde=True)
plt.title("Budget Distribution by Order Type")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 7: Top 10 highest budgets

top_10_budgets = data.nlargest(10, 'budget')[['order_id', 'budget']]
print("Top 10 highest budgets:\n", top_10_budgets)
```

```python
# Question 8: Median budget

median_budget = data['budget'].median()
print("Median budget:", median_budget)
```

```python
# Question 9: Budget range for each type of order

budget_range = data.groupby('type_of_order')['budget'].agg(['min', 'max'])
print("Budget range per order type:\n", budget_range)
```

```python
# Question 10: Count of orders in different budget ranges

bins = [0, 2000, 4000, float('inf')]
labels = ['<2000', '2000-4000', '>4000']
data['budget_range'] = pd.cut(data['budget'], bins=bins, labels=labels)
budget_range_count = data['budget_range'].value_counts()
print("Orders in different budget ranges:\n", budget_range_count)
```

```python
# Question 11: What is the distribution of the budget for each type_of_order (Delivery vs Takeaway) across different order_id?

plt.figure(figsize=(10, 6))
sns.violinplot(x='type_of_order', y='budget', data=data, palette='Set2')
plt.title("Distribution of Budget for Delivery vs Takeaway Orders")
plt.xlabel("Order Type")
plt.ylabel("Budget")
plt.show()
```

```python
# Question 12: Average budget spent by each customer

avg_budget_customer = data.groupby('customer_id')['budget'].mean()
print("Average budget per customer:\n", avg_budget_customer)
```

```python
# Question 13: Customer with highest budget

highest_spending_customer = data.groupby('customer_id')['budget'].sum().idxmax()
print("Customer with highest budget:", highest_spending_customer)
```

```python
# Question 14: Popularity of 'Delivery' vs 'Takeaway' orders

order_popularity = data['type_of_order'].value_counts()
print("Order popularity:\n", order_popularity)
```

```python
# Question 15: Average budget for customers who prefer 'Delivery'

delivery_avg_budget = data[data['type_of_order'] == 'Delivery']['budget'].mean()
print("Average budget for Delivery orders:", delivery_avg_budget)
```

```python
# Question 16: Distribution of budgets for 'Delivery' orders

delivery_data = data[data['type_of_order'] == 'Delivery']
plt.figure(figsize=(10, 6))
sns.histplot(delivery_data['budget'], kde=True, color='blue')
plt.title("Budget Distribution for Delivery Orders")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 17: Distribution of budgets for 'Takeaway' orders

takeaway_data = data[data['type_of_order'] == 'Takeaway']
plt.figure(figsize=(10, 6))
sns.histplot(takeaway_data['budget'], kde=True, color='green')
plt.title("Budget Distribution for Takeaway Orders")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 18: Outliers in the budget column

plt.figure(figsize=(8, 6))
sns.boxplot(data['budget'])
plt.title("Budget Outliers")
plt.xlabel("Budget")
plt.show()
```

```python
# Question 19: Relationship between order type and budget

plt.figure(figsize=(8, 6))
sns.boxplot(x='type_of_order', y='budget', data=data)
plt.title("Budget by Order Type")
plt.show()
```

```python
# Question 20: How order type affects average budget

avg_budget_by_type = data.groupby('type_of_order')['budget'].mean()
print("Average budget by order type:\n", avg_budget_by_type)
```

```python
# Question 21: What is the distribution of order budgets for 'Delivery' and 'Takeaway' orders?

plt.figure(figsize=(8, 6))
sns.boxplot(x='type_of_order', y='budget', data=data, palette='pastel')
plt.title("Budget Distribution by Order Type (Delivery vs Takeaway)")
plt.xlabel("Order Type")
plt.ylabel("Budget")
plt.show()
```

```python
# Question 22: Distribution of the budget (histogram or boxplot)

plt.figure(figsize=(10, 6))
sns.histplot(data['budget'], kde=True)
plt.title("Budget Distribution")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 23: Average budget per customer type

avg_budget_type = data.groupby('type_of_order')['budget'].mean()
print("Average budget per customer type:\n", avg_budget_type)
```

```python
# Question 24: Bar plot for the number of orders per customer type

plt.figure(figsize=(8, 6))
sns.countplot(x='type_of_order', data=data)
plt.title("Number of Orders by Type")
plt.xlabel("Order Type")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Question 25: Relationship between 'order_id' and 'budget' over time

avg_budget_by_order_type = data.groupby('type_of_order')['budget'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=avg_budget_by_order_type.index, y=avg_budget_by_order_type.values, palette='Set2')
plt.title("Average Budget by Order Type (Delivery vs Takeaway)")
plt.xlabel("Order Type")
plt.ylabel("Average Budget")
plt.show()


```

```python
# Question 25: Relationship between 'order_id' and 'budget' over time

data_sorted = data.sort_values('order_id')

plt.figure(figsize=(12, 6))
sns.lineplot(x=data_sorted.index, y='budget', data=data_sorted, marker='o', color='blue')
plt.title("Relationship Between Order Sequence and Budget")
plt.xlabel("Order ID (Simulated Time)")
plt.ylabel("Budget")
plt.show()
```

```python
# Question 26: Percentage of 'Delivery' vs 'Takeaway' orders

order_percent = data['type_of_order'].value_counts(normalize=True) * 100
print("Percentage of orders by type:\n", order_percent)
```

```python
# Question 27: What is the total budget for each order type (Delivery and Takeaway)?

total_budget_by_order_type = data.groupby('type_of_order')['budget'].sum()

plt.figure(figsize=(8, 6))
total_budget_by_order_type.plot(kind='bar', color=['blue', 'green'])
plt.title("Total Budget for Each Order Type (Delivery vs Takeaway)")
plt.xlabel("Order Type")
plt.ylabel("Total Budget")
plt.xticks(rotation=0)
plt.show()
```

```python
# Question 28: Budget spending trend for 'Delivery' vs 'Takeaway' orders over time

# Simulate 'month' and calculate the average budget by 'month' and 'type_of_order'
data['month'] = np.random.choice(range(1, 13), size=len(data))

# Plotting the budget spending trend by order type
data.groupby(['month', 'type_of_order'])['budget'].mean().unstack().plot(kind='line', figsize=(10, 6), marker='o')
plt.title("Budget Spending Trend by Order Type")
plt.xlabel("Month")
plt.ylabel("Average Budget")
plt.xticks(range(1, 13))
plt.show()
```

```python
# Question 29: Average order value by customer segment (Delivery vs Takeaway)

avg_order_value = data.groupby('type_of_order')['budget'].mean()
print("Average order value by customer segment:\n", avg_order_value)
```

```python
# Question 30: Order type with higher budgets

order_type_budget = data.groupby('type_of_order')['budget'].sum()
print("Total budget by order type:\n", order_type_budget)
```

```python
#31.Question: How do customers' spending patterns differ between 'Delivery' and 'Takeaway' orders across different budget ranges?

bins = [0, 2000, 4000, float('inf')]
labels = ['<2000', '2000-4000', '>4000']
data['budget_range'] = pd.cut(data['budget'], bins=bins, labels=labels)

order_distribution = data.groupby(['type_of_order', 'budget_range']).size().unstack()


plt.figure(figsize=(10, 6))
order_distribution.plot(kind='bar', stacked=True, color=sns.color_palette("Set2", n_colors=3))
plt.title("Customer Spending Patterns by Order Type and Budget Range")
plt.xlabel("Order Type")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.legend(title="Budget Range", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
