```python
#1. 1. What is the total number of customers in the dataset?  

total_customers = data['customer_id'].nunique()
print("Total number of customers :",total_customers)
```

```python
#2. What is the total number of orders?

total_order = data['order_'].count()
print("Total number of orders :",total_order)
```

```python
# 3. How many unique products are there in the dataset?

unique_products = data['product'].nunique()
print("Unique products:", unique_products)
```

```python
# 4. Which product has the highest number of orders?

highest_ordered_product = data['product'].value_counts().idxmax()
print("Most ordered product:", highest_ordered_product)
```

```python
# 5. What is the total sum of tips?

total_tips = data['tips'].sum()
print("Total sum of tips:", total_tips)
```

```python
# 6. What is the average tip amount?

average_tip = data['tips'].mean()
print("Average tip amount:", average_tip)
```

```python
# 7. What is the highest tip given?

highest_tip = data['tips'].max()
print("Highest tip given:", highest_tip)
```

```python
# 8. What is the lowest tip given?

lowest_tip = data['tips'].min()
print("Lowest tip given:", lowest_tip)
```

```python
# 9. Which customer gave the highest tip?

highest_tip_customer = data.loc[data['tips'].idxmax(), 'customer_name']
print("Customer with highest tip:", highest_tip_customer)
```

```python
# 10. How many customers have loyalty points greater than 100?

loyal_customers = data[data['loyalty_points'] > 100].shape[0]
print("Customers with loyalty points > 100:", loyal_customers)
```

```python
# 11. What is the average number of loyalty points?

average_loyalty_points = data['loyalty_points'].mean()
print("Average loyalty points:", average_loyalty_points)
```

```python
# 12. Which product generated the highest tips on average?

highest_tips_product = data.groupby('product')['tips'].mean().idxmax()
print("Product with highest average tips:", highest_tips_product)
```

```python
# 13. How many customers provided feedback mentioning "sweet"?

sweet_feedback_count = data[data['feedback'].str.contains("sweet")].shape[0]
print("Feedback mentioning 'sweet':", sweet_feedback_count)
```

```python
# 14. What is the most common feedback provided by customers?

most_common_feedback = data['feedback'].mode()[0]
print("Most common feedback:", most_common_feedback)
```

```python
# 15. Which product received the most positive feedback?

pos_feedback_product = data[data['feedback'].str.contains("good|excellent|great", case=False, na=False)]['product'].value_counts().idxmax()
print("Product with most positive feedback:", pos_feedback_product)
```

```python
# 16. How many customers placed more than one order?

repeat_customers = data['customer_id'].value_counts()[data['customer_id'].value_counts() > 1].count()
print("Customers with more than one order:", repeat_customers)
```

```python
# 17. What is the most common domain in customer emails?

email_domains = data['email'].str.split('@').str[1].value_counts().idxmax()
print("Most common email domain:", email_domains)
```

```python
# 18. How many unique email domains are in the dataset?

unique_email_domains = data['email'].str.split('@').str[1].nunique()
print("Unique email domains:", unique_email_domains)
```

```python
# 19. Which customer has the lowest number of loyalty points?

lowest_loyalty_customer = data.loc[data['loyalty_points'].idxmin(), 'customer_name']
print("Customer with lowest loyalty points:", lowest_loyalty_customer)
```

```python
# 20. How many customers have their phone numbers ending in 0?

data['customer_phonenumber'] = data['customer_phonenumber'].astype(str)  # Convert to string
phone_ending_zero = data[data['customer_phonenumber'].str.endswith('0')].shape[0]
print("Customers with phone numbers ending in 0:", phone_ending_zero)
```

```python
# 21. What is the distribution of loyalty points?

plt.figure(figsize=(10, 6))
sns.histplot(data['loyalty_points'], kde=True, bins=30, color='blue')
plt.title('Distribution of Loyalty Points')
plt.xlabel('Loyalty Points')
plt.ylabel('Frequency')
plt.show()
```

```python
# 22. Plot a histogram of tip amounts to visualize their distribution.

plt.figure(figsize=(10, 6))
sns.histplot(data['tips'], kde=True, bins=30, color='green')
plt.title('Distribution of Tips')
plt.xlabel('Tips')
plt.ylabel('Frequency')
plt.show()
```

```python
# 23. Create a bar chart showing the top 10 most ordered products.

top_products = data['product'].value_counts().head(10)
top_products.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('Top 10 Most Ordered Products')
plt.xlabel('Product')
plt.ylabel('Order Count')
plt.show()
```

```python
# 24. Create a pie chart representing the percentage distribution of orders by product categories.

category_distribution = data['product'].value_counts().head(5)
category_distribution.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Orders by Product Categories')
plt.ylabel('')
plt.show()
```

```python
# 25. Generate a heatmap to visualize the correlation between numerical columns in the dataset.

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 26. Create a scatter plot showing the relationship between loyalty points and tips.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='loyalty_points', y='tips', data=data)
plt.title('Loyalty Points vs Tips')
plt.xlabel('Loyalty Points')
plt.ylabel('Tips')
plt.show()
```

```python
# 27. Create a boxplot to compare tips across different product categories.

plt.figure(figsize=(12, 6))
sns.boxplot(x='product', y='tips', data=data)
plt.title('Tips by Product Categories')
plt.xlabel('Product')
plt.ylabel('Tips')
plt.xticks(rotation=45)
plt.show()
```

```python
# 28. Plot the count of orders by customer using a horizontal bar chart for the top 10 customers.

top_customers = data['customer_id'].value_counts().head(10)
top_customers.plot(kind='barh', figsize=(10, 6), color='purple')
plt.title('Top 10 Customers by Order Count')
plt.xlabel('Order Count')
plt.ylabel('Customer ID')
plt.show()
```

```python
from wordcloud import WordCloud

# 29. Create a word cloud of the most frequent words in the feedback column.

feedback_text = ' '.join(data['feedback'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(feedback_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Feedback')
plt.show()
```

```python
# 30. Plot a KDE (Kernel Density Estimate) plot to visualize the density of loyalty points.

plt.figure(figsize=(10, 6))
sns.kdeplot(data['loyalty_points'], shade=True, color='red')
plt.title('Density Plot of Loyalty Points')
plt.xlabel('Loyalty Points')
plt.ylabel('Density')
plt.show()
```
